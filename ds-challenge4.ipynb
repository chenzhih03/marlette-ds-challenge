## Encode the categorical features

```python
def data_cleaning(df, cols):
    '''Use one-hot encoding to encode all the categorical features. 
    '''
    l = [df]
    for col in cols:
        df_dummies = pd.get_dummies(df[col], prefix=col).iloc[:,:-1]
        l.append(df_dummies)
       
    df = pd.concat(l, axis=1).drop(cols, axis=1)
    
    return(df)
```

```python
Cols_cat_names = ['cat1','cat2','cat3','cat4','cat5']
df_cleaned = data_cleaning(df, Cols_cat_names)
```

```python
df_cleaned.head()
```

```python
X=df_cleaned.drop('target', axis=1)
y = df_cleaned['target']
```

### Oversampling the minority class and downsampling the majority class.

```python
from imblearn.over_sampling import SMOTE
from imblearn.under_sampling import RandomUnderSampler
from imblearn.pipeline import Pipeline
```

```python
over_sampling = SMOTE(sampling_strategy=0.5)
under_sampling = RandomUnderSampler(sampling_strategy=1)
```

## Define a function to tune the parameters of models

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, roc_auc_score
from hyperopt import hp, fmin, tpe, STATUS_OK, Trials
```

```python
def optimize(params, model_init, x, y, score_method='roc_auc_score'):
    model = model_init(**params)
    kf = StratifiedKFold(n_splits=5)
    score = []
    for idx in kf.split(X=x,y=y):
        train_idx, validation_idx = idx[0], idx[1]
        x_train = x[train_idx]
        y_train = y[train_idx]
        
        y_validation = y[validation_idx]
        x_validation = x[validation_idx]
        
        pipeline = Pipeline(steps=[('over_sampling', over_sampling), \
            ('under_sampling', under_sampling),\
                ('model', model)])
        #pipeline.fit_resample(x_train, y_train)
        pipeline.fit(x_train, y_train)
        model = pipeline['model']
        y_predict = model.predict(x_validation)
        if score_method == 'roc_auc_score':
            auc_score = roc_auc_score(y_validation, model.predict_proba(x_validation)[:,1])
            score.append(auc_score)
        if score_method == 'accuracy_score':
            acc_score = accuracy_score(y_validation, y_predict)
            score.append(acc_score)
        if score_method == 'precision_score':
            pre_score = precision_score(y_validation, y_predict)
            score.append(pre_score)
        if score_method == 'recall_score':
            rec_score = recall_score(y_validation, y_predict)
            score.append(rec_score)
            
    mean_score = np.mean(score)
    return {'loss': -1*mean_score, 'status':STATUS_OK}
        
        
```

### Modeling

LogisticRegression

```python
from functools import partial
import warnings
from sklearn.exceptions import ConvergenceWarning
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
param_space = {
    'C': hp.uniform('C', 0, 10),
    # 'class_weight': 'balanced',#hp.choice('class_weight',  ['balanced']),#[{1:2},{1:3}, {1:4},{1:5},{1:7}, {1:10}, {1:20}, {1:40}]),
    'max_iter':500
}
```

```python
optimize_lr = partial(optimize, model_init = LogisticRegression,x=X_train_final, y = y_train.values, score_method='accuracy_score')
```

```python
trials = Trials()
```

```python
with warnings.catch_warnings():
    warnings.simplefilter("ignore", category=ConvergenceWarning)
    best_lr =  fmin(
        fn = optimize_lr,
        space = param_space,
        algo = tpe.suggest,
        max_evals = 50,
        verbose=0,
        trials = trials
        )
```
